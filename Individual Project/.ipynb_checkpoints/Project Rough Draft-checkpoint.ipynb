```python
list_of_dfs = []
```

```python
for i in np.arange(1997,2014):
    new_df = pd.read_excel('DIGHUM data/Excel data/national_' + str(i) + '_dl.xls')
    new_df.columns = [x.lower() for x in new_df.columns]
    columns = new_df.columns
    new_col = []
    for i in np.arange(len(columns)):
        temp = columns[i].replace('w','')
        new_col.append(temp)
    new_df.columns = new_col
    new_df.select('a_median', 'a_pct90')
    list_of_dfs.append(new_df)
```

```python
print(len(list_of_dfs))
```

```python
big_table = pd.concat(list_of_dfs)
```
